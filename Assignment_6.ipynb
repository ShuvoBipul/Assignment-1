```python
gymmembers = pd.read_csv('/content/drive/MyDrive/SkillMorph_Dataset/gym_membership.csv')  # No header in raw CSV
```

```python
# Check data
print(gymmembers.head())
print(gymmembers.shape)  # How many rows and columns
print(gymmembers.columns.tolist())  # Column names
```

```python
print("\nDataFrame Description:")
print(gymmembers.describe())
```

```python
# Basic DataFrame information
print("\nDataFrame Info:")
print(gymmembers.info())
```

```python


# Assuming 'gym' is your loaded DataFrame (e.g., gym = pd.read_csv('gym_membership.csv'))
counts = gymmembers['abonoment_type'].value_counts()

plt.figure(figsize=(8, 6))
counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Gym Membership Types Distribution')
plt.xlabel('Abonoment Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
```

```python


plt.figure(figsize=(10, 6))
plt.hist(gymmembers['avg_time_in_gym'], bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Distribution of Average Time in Gym')
plt.xlabel('Average Time in Gym (minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
```

```python

plt.scatter(gymmembers['Age'], gymmembers['visit_per_week'])
plt.grid(True)
plt.xlabel('Age')
plt.ylabel('Visits per week')
plt.title('Age vs Visits per week')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
```

```python
# Assuming 'gymmembers' DataFrame is loaded
plt.figure(figsize=(8, 6))
sns.boxplot(data=gymmembers, x='gender', y='Age', palette='Set2')
plt.title('Age Distribution by Gender')
plt.ylabel('Age')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
```

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
```

```python


# Assuming 'gym' DataFrame is loaded
plt.figure(figsize=(10, 6))
sns.countplot(data=gymmembers, x='days_per_week', palette='viridis')
plt.title('Distribution of Gym Visits by Days Per Week')
plt.xlabel('Days Per Week')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=gymmembers, x='abonoment_type', y='visit_per_week', palette='Set3', capsize=0.1)
plt.title('Average Visits Per Week by Membership Type')
plt.ylabel('Average Visits Per Week')
plt.xlabel('Membership Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
